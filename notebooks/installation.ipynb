# Installation

## conda / mamba environment

First, install the **mamba** environment using the `core-env.yml` file from the main pLAST repository:  
`mamba env create -n plast -f core-env.yml`

You can also use **conda**:  
`conda env create -n plast -f core-env.yml`

Alternatively, install the following dependencies manually:
- `hmmer >= 3.4`
- `mmseqs2 >= 17.b804f`
- `prodigal >= 2.6.3`

Once all required packages are installed, activate the environment:  
`mamba activate plast`  
or  
`conda activate plast`

If you plan to use Jupyter notebooks, install the `ipykernel` package as well:  
`mamba install ipykernel`  
or  
`conda install ipykernel`
